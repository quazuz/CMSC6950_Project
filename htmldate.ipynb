```python
from htmldate import find_date
```

```python
import pandas as pd
import numpy as np 
```

```python
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
```

```python
import csv
df = pd.read_csv("dates.csv")
np_url = df['url'].values.tolist()
#print(df)
print(np_url)

dates = []
for url in np_url:
    date = find_date(url)
    dates.append(date)
    #print(date)
    for date in dates:
        print(date)
```

## Getting url names

```python
from urllib.request import urlopen
from lxml.html import parse

url = 'http://news.mit.edu/2020/ideastream-showcases-breakthrough-technologies-across-mit-0624'
page = urlopen(url)
p = parse(page)
print (p.find(".//title").text)
```

```python
column = widgets.Dropdown(options=list(df['url']),description='Web Link')
ui = widgets.HBox([column])
def print_date(column):
    date = find_date(column)
    print(date)

out = widgets.interactive_output(print_date, {'column': column})

display(ui, out)
```

```python
# df.to_csv('dates.csv')   
import matplotlib
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(dates, np_url)

ax.set(ylabel='URL', xlabel='Dates',
       title='PLOT SHOWING PUBLICATION DATES OF DIFFERENT WEB ARTICLES')
ax.grid()

fig.savefig("test.png")
plt.show()
```
